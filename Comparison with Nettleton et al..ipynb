```python
from Benchmark.protocol import *
```

```python
p = Protocol(10, 2, [0.0, 0.05, 0.1, 0.20, 0.25, 0.33, 0.5, 0.66, 0.75, 0.9, 1.0, 1.25], {"regular" : lambda x : x, "onehot" : lambda x : pd.get_dummies(x)})
```

```python
p.splitDataset("breastcancer_full", 34.4, n_jobs=10)
```

```python
# FULL TESTS
n_trees = 100
n_jobs = 3
xgbparams = {"nthread": 1, "eta" : 0.3, "min_child_weight" : 0, "lambda": 1, "tree_method" : "auto"}
toZeroOne = lambda ys : (ys.copy()+1.0)/2
```

```python
skm = {
     "regular": {
        "khiops" : sklearnFit(PyKhiopsClassifier, evaluateProbaKhiops),
        "khiopsRF" : sklearnFit(PyKhiopsClassifier, evaluateProbaKhiops, n_trees=n_trees)
    },
    "onehot" : {
        "skRF" : sklearnFit(RandomForestClassifier, evaluateProba,
            {"n_estimators": n_trees, "n_jobs": 1}),
        "logisticRegression" : sklearnFit(LogisticRegression, evaluateScore,
            {"solver" : "lbfgs", "max_iter" : 20000, "n_jobs": 1}),
        "linearSVC" : sklearnFit(LinearSVC, evaluateScore,
            {"max_iter" : 20000, "dual" : False}),
        "XGB_SQUERR" : xgbFit('reg:squarederror', n_trees, params=xgbparams),
        "XGB_HINGE" : xgbFit('binary:hinge', n_trees, params=xgbparams, prepy = toZeroOne),
        "XGB_UNHINGED" : xgbFit(buildUnhinged(), n_trees, params=xgbparams, prepy = toZeroOne),
        "XGB_RAMP" : xgbFit(buildModifiedRamp(ep = 1e-16, r = 0.5), n_trees, params=xgbparams, prepy = toZeroOne)
    }
}

# Ordered by ascending cells amount
for i in range(p.K*p.R):
    p.runModels("breastcancer_full", skm, i, n_jobs=n_jobs)
```

```python
wekam = {
    "regular": {
        "wekaRF" : wekaFit(params={"n_estimators" : n_trees, "n_jobs" : n_jobs})
    }
}

# Ordered by ascending cells amount
for i in range(p.K*p.R):
    #weka runs cannot be parralelized. we can only count on the weka parralel implementation of RF
    p.runModels("breastcancer_full", wekam, i, n_jobs=1)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
from adjustText import adjust_text
sns.set(style="whitegrid")
```

```python
res = pd.read_csv("Results/results_breastcancer_full_2x10x12.csv")
res
```

```python
palette ={"khiops":"darkorange", "khiopsRF":"darkorange","skRF":"cornflowerblue","logisticRegression":"limegreen", "linearSVC":"limegreen", "XGB_SQUERR":"brown", "XGB_HINGE":"brown", "XGB_RAMP":"brown", "XGB_UNHINGED":"brown", "wekaRF":"cornflowerblue"}

plt.figure(figsize=(15, 10))

margin = .25
ep = 0.01
metric = "test_bacc"
rho_min, rho_max = 0.0, 1.25

ax = sns.lineplot(x="rho", y=metric, hue="model", marker="o", data=res, ci=None, palette=palette)
ax.set_xlim(0-margin, 1.25+margin)

mean = res.groupby(["rho", "model"]).mean()
lefttexts, righttexts = [], []
for m, mcol in palette.items():
    textprop = {"color":mcol, "verticalalignment":"center", "fontweight": "semibold"}
    lefttexts.append(ax.text(rho_max+ep, mean.T[rho_max, m].T[metric]+(random.random()/10000), m, **textprop))
    righttexts.append(ax.text(rho_min-ep, mean.T[rho_min, m].T[metric]+(random.random()/10000), m, **textprop))
adjust_params = {"only_move":{'points':'xy', 'text':'y', 'objects':'xy'}, "autoalign":False, "force_text" : 0.4 }
adjust_text(lefttexts, ha="left", **adjust_params)
adjust_text(righttexts, ha="right",**adjust_params)
ax.set_xticks([rho for rho in res["rho"].unique() if rho_min <= rho <= rho_max ])

pass
```
